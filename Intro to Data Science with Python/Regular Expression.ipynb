```python
#Creating a re match function to return match pbject or none on failure
import re

result=re.match('Analytics', "Analytics and the science of Data is interesting")
print(result)

result2 = re.match('Largest', 'Analytics and the science of Data is interesting')
print(result2)
```

```python
print(result.group())
```

```python
#Re search go through the entire expression looking for the index
result3=re.search('Analytics', "I love Data Science and Analytics")
print(result3.group())
```

```python
#re findall() is m,ore efficient and it returns all the occurence of the pattern from the entire expression
result4 = re.findall('Analytics', 'I love Data Science and Analytics, especially the Analytics part')
print(result4)
```

**SPECIAL SEQUENCE**

```python
#\b return all item with same suffix

str1 = 'I love Data Science and Analytics, especially the Analytics part'

x = re.findall("ics\b", str1)

print(x)
```

```python
#\d picks any digit in an expression
str2 = "2 million people were in that space in year 2021"

y = re.findall('\d+', str2)

print(y)

if(y):
  print(f'we have {len(y)} in the expression')
else:
  print('No match')
```

```python
#\D return match without any digit in an expression
str = "2 million people were in that space in year 2021"

y = re.findall('\D+', str)

print(y)
```
