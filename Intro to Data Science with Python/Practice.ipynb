```python
#Imort Necessary Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LinearRegression
```

```python
#Load the data

data = pd.read_csv(path+'Simple linear regression.csv')

data.head()
```

```python
#Declare the dependent and independent variables
y = data['GPA']
x = data['SAT']
```

```python
x.shape
```

```python
y.shape
```

```python
x_matrix = x.values.reshape(-1,1)

x_matrix.shape
```

```python
#Regression itself

reg = LinearRegression()

reg.fit(x_matrix,y)
```

```python
#R_Squared

reg.score(x_matrix,y)
```

```python
#Coefficient
reg.coef_
```

```python
#Intercept
reg.intercept_
```

```python
#predictions
new_dat = pd.DataFrame(data=[1740,1760],columns=['SAT'])
new_dat
```

```python
reg.predict(new_dat)
```

```python
new_dat['Predicted_GPA'] = reg.predict(new_dat)
new_dat
```

```python
plt.scatter(x,y)
yh = reg.coef_*x_matrix+reg.intercept_
fig = plt.plot(x, yh, lw=4, c="orange", label='regression')
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```
