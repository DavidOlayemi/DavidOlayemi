```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style='darkgrid')
```

```python
data = pd.read_csv(path+'bigmart_data.csv')

data = data.dropna(how='any')

data['Visibility_Scaled']=data['Item_Visibility']*100
```

```python
sns.lineplot(x='Item_Weight', y='Item_MRP', data=data[:50])
```

```python
sns.barplot(x='Item_Type', y='Item_MRP', data = data[:5])
```

```python
#Histogram
sns.distplot(data['Item_MRP'])
```

```python
q = data['Item_Outlet_Sales'].quantile(0.99)
data_new = data[data['Item_Outlet_Sales']<q]
#Box plot
sns.boxplot(data_new['Item_Outlet_Sales'], orient='vertical')
```

```python
#Violin plot

sns.violinplot(data['Item_Outlet_Sales'], orient='vertical', color='magenta')
```

```python
#scatter
sns.scatterplot(x='Item_MRP', y='Item_Outlet_Sales', data=data[:200])
```

```python
#hue semantic
sns.relplot(x='Item_MRP', y='Item_Outlet_Sales', kind='scatter', hue='Item_Type', data=data[:200])
```

```python
sns.relplot(x='Item_Weight', y='Item_MRP', kind='line', hue='Item_Type', data=data[:50])
```

```python
sns.relplot(x='Item_MRP', y='Item_Outlet_Sales', hue='Visibility_Scaled', size='Visibility_Scaled', data=data[:50])
```

```python
sns.relplot(x='Item_Weight', y='Item_Visibility',hue='Outlet_Size',style='Outlet_Size',col='Outlet_Size', data=data[:100])
```

```python
#Categorical Plots
#Strip Plot
sns.catplot(x='Outlet_Size', y='Item_Outlet_Sales', kind='strip', data=data[:250]);
```

```python
#Swarm plot
sns.catplot(x='Outlet_Size', y='Item_Outlet_Sales', kind='swarm', data=data[:250]);
```

```python
#Box plot
sns.catplot(x='Outlet_Size', y='Item_Outlet_Sales', kind='box', data=data);
```

```python
#Violin plot
sns.catplot(x='Outlet_Size', y='Item_Outlet_Sales', kind='violin', data=data);
```

```python
#Density Plot
#distribution of Item Visibility
plt.figure(figsize=(7,7))
sns.kdeplot(data['Item_Visibility'], shade=True)
```

```python
sns.kdeplot(data['Item_MRP'], shade=True)
```

```python
#Density and Histogram
plt.figure(figsize=(7,7))
sns.displot(data['Item_Outlet_Sales']);
```
